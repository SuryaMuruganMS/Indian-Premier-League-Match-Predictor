```python
import pandas as pd
import numpy as np
import re
import random
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx

raw_data_2024 = """Match Date\tTime\tMatch\tToss Winner\tMatch Winner and Toss
22- Mar- 2024\t7:30 PM\tRCB vs CSK\tRCB\tCSK Won by 6 Wickets
23- Mar- 2024\t3:30PM\tDC vs PBKS\tPBKS\tPBKS Won by 4 Wickets
23- Mar- 2024\t7:30 PM\tKKR vs SRH\tSRH\tKKR Won by 4 Runs
24- Mar- 2024\t3:30PM\tRR vs LSG\tRR\tRR Won by 20 Runs
24- Mar- 2024\t7:30 PM\tGT vs MI\tMI\tGT Won by 6 Runs
25- Mar- 2024\t7:30 PM\tPBKS vs RCB\tRCB\tRCB Won by 4 Wickets
26- Mar- 2024\t7:30 PM\tCSK vs GT\tGT\tCSK Won by 63 Runs
27- Mar- 2024\t7:30 PM\tSRH vs MI\tMI\tSRH Won by 31 Runs
28- Mar- 2024\t7:30 PM\tRR vs DC\tDC\tRR Won by 12 Runs
29- Mar- 2024\t7:30 PM\tRCB vs KKR\tKKR\tKKR Won by 7 Wickets
30- Mar- 2024\t7:30 PM\tLSG vs PBKS\tLSG\tLSG Won by 21 Runs
31- Mar- 2024\t3:30PM\tGT vs SRH\tSRH\tGT Won by 7 Wickets
31- Mar- 2024\t7:30 PM\tDC vs CSK\tDC\tDC Won by 20 Runs
1- April- 2024\t7:30 PM\tMI vs RR\tRR\tRR Won by 6 Wickets
2- April- 2024\t7:30 PM\tRCB vs LSG\tRCB\tLSG Won by 28 Runs
3- April- 2024\t7:30 PM\tDC vs KKR\tKKR\tKKR Won by 106 Runs
4- April-2024\t7:30 PM\tGT vs PBKS\tPBKS\tPBKS won by 3 wickets
5-April-2024\t7:30 PM\tSRH vs CSK\tSRH\tSRH won by 6 wickets
6- April- 2024\t7:30 PM\tRCB vs RR\tRR\tRR Won by 6 Wickets
7- April-2024\t7:30 PM\tMI vs DC\tDC\tMI won by 29 runs
7 -April-2024\t7:30 PM\tLSG vs GT\tLSG\tLSG won by 33 runs
8-April-2024\t7:30 PM\tCSK vs KKR\tCSK\tCSK won by 7 wickets
9-April-2024\t7:30 PM\tPBKS vs SRH\tPBKS\tSRH won by 2 runs
10-April-2024\t7:30 PM\tRR vs GT\tGT\tGT won by 3 wickets
11-April-2024\t7:30 PM\tMI vs RCB\tMI\tMI won by 7 wickets
12-April-2024\t7:30 PM\tLSG vs DC\tLSG\tDC won by 6 wickets
13-April-2024\t7:30 PM\tPBKS vs RR\tRR\tRR won by 3 wickets
14-April-2024\t3:30 PM\tKKR vs LSG\tKKR\tKKR won by 8 wickets
14-April-2024\t7:30 PM\tMI vs CSK\tMI\tCSK won by 20 runs
15-April-2024\t7:30 PM\tRCB vs SRH\tRCB\tSRH won by 25 runs
16-April-2024\t7:30 PM\tKKR vs RR\tRR\tRR won by 2 Wickets
17-April-2024\t7:30 PM\tGT vs DC\tDC\tDC won by 6 wickets
18-April-2024\t7:30 PM\tPBKS vs MI\tPBKS\tMI won by 9 runs
19-April-2024\t7:30 PM\tLSG vs CSK\tLSG\tLSG won by 8 wickets
20-April-2024\t7:30 PM\tDC vs SRH\tDC\tSRH won by 67 Runs
21-April-2024\t3:30 PM\tKKR vs RCB\tRCB\tKKR won by 1 Run
21-April-2024\t7:30 PM\tPBKS vs GT\tPBKS\tGT won by 3 wickets
22 April 2024\t7:30 PM\tRR vs MI\tRR\tRR won by 9 Wickets
23 April 2024\t7:30 PM\tCSK vs LSG\tLSG\tLSG won by 6 wickets
24 April-2024\t7:30 PM\tDC vs GT\tDC\tDC won by 4 Runs
25 April-2024\t7:30 PM\tSRH vs RCB\tRCB\tRCB won by 35 Runs
26 April-2024\t7:30 PM\tKKR vs PBKS\tPBKS\tPBKS won by 8 wickets
27 April-2024\t3:30 PM\tDC vs MI\tMI\tDC won by 10 runs
27 April-2024\t7:30 PM\tLSG vs RR\tRR\tRR won by 7 wickets
28 April-2024\t3:30 PM\tGT vs RCB\tRCB\tRCB won by 9 wickets
28 April-2024\t7:30 PM\tCSK vs SRH\tSRH\tCSK won by 78 runs
29 April-2024\t7:30 PM\tKKR vs DC\tDC\tKKR won by 7 wickets 
30 April-2024\t7:30 PM\tLSG vs MI\tLSG\tLSG won by 4 wickets
1 May-2024\t7:30 PM\tCSK vs PBKS\tPBKS\tPBKS won by 7 wickets
2 May-2024\t7:30 PM\tSRH vs RR\tRR\tSRH won by 1 run
3 May-2024\t7:30 PM\tKKR vs MI\tMI\tKKR won by 24 runs
4 May-2024\t7:30 PM\tRCB vs GT\tRCB\tRCB won by 4 wickets
5 May-2024\t3:30 PM\tPBKS vs CSK\tPBKS\tCSK won by 28 runs
5 May-2024\t7:30 PM\tLSG vs KKR\tLSG\tKKR won by 98 runs
6 May-2024\t7:30 PM\tMI vs SRH\tMI\tMI won by 7 wickets
7 May-2024\t7:30 PM\tDC vs RR\tRR\tDC won by 20 runs
8 May-2024\t7:30 PM\tSRH vs LSG\tLSG\tSRH won by 10 wickets
9 May-2024\t7:30 PM\tRCB vs PBKS\tPBKS\tRCB won by 60 runs
10 May-2024\t7:30 PM\tGT vs CSK\tCSK\tGT won by 35 runs
11 May-2024\t7:30 PM\tKKR vs MI\tMI\tKKR won by 18 runs
12 May-2024\t3:30 PM\tCSK vs RR\tRR\tCSK won by 5 wickets 
12 May-2024\t7:30 PM\tRCB vs DC\tDC\tRCB won by 47 runs
13 May-2024\t7:30 PM\tGT vs KKR\t\tMatch abandoned without toss
14 May-2024\t7:30 PM\tDC vs LSG\tLSG\tDC won by 19 runs
15 May-2024\t7:30 PM\tPBKS vs RR\tRR\tPBKS won by 5 wickets
16 May-2024\t7:30 PM\tSRH vs GT\t\tMatch abandoned without toss
17 May-2024\t7:30 PM\tLSG vs GT\tMI\tLSG won by 18 runs
18 May-2024\t7:30 PM\tRCB vs CSK\tCSK\tRCB won by 27 runs
19 May-2024\t3:30 PM\tPBKS vs SRH\tPBKS\tSRH won by 4 wickets
19 May-2024\t7:30 PM\tRR vs KKR\tKKR\tAbandoned with a toss
21 May-2024\t7:30 PM\tKKR vs SRH\tSRH\tKolkata Knight Riders won by 8 wkts
22 May-2024\t7:30 PM\tRCB vs RR\tRR\tRajasthan Royals won by 4 wkts
24 May-2024\t7:30 PM\tRR vs SRH\tRR\tSunrisers Hyderabad won by 36 runs
26 May-2024\t7:30 PM\tKKR vs SRH\tSRH\tKKR won by 8 wickets
"""

df_raw = pd.read_csv(pd.io.common.StringIO(raw_data_2024), sep="\t")

def extract_teams(match_str):
    parts = match_str.split("vs")
    if len(parts) == 2:
        return parts[0].strip(), parts[1].strip()
    return None, None

abbr_mapping = {
    "CSK": "Chennai Super Kings",
    "RCB": "Royal Challengers Bangalore",
    "DC": "Delhi Capitals",
    "PBKS": "Punjab Kings",
    "KKR": "Kolkata Knight Riders",
    "SRH": "Sunrisers Hyderabad",
    "RR": "Rajasthan Royals",
    "LSG": "Lucknow Super Giants",
    "GT": "Gujarat Titans",
    "MI": "Mumbai Indians"
}

def extract_actual_winner(txt):
    if pd.isnull(txt) or "abandoned" in txt.lower():
        return None
    m = re.search(r"^(\S+)\s+won", txt.lower())
    if m:
        code = m.group(1).upper()
        return abbr_mapping.get(code, code)
    return None

df_raw[['team1_abbr','team2_abbr']] = df_raw['Match'].apply(lambda x: pd.Series(extract_teams(x)))
df_raw['team1'] = df_raw['team1_abbr'].map(abbr_mapping)
df_raw['team2'] = df_raw['team2_abbr'].map(abbr_mapping)
df_raw['toss_winner'] = df_raw['Toss Winner'].map(abbr_mapping)
df_raw['actual_winner'] = df_raw['Match Winner and Toss'].apply(extract_actual_winner)

def guess_toss_decision(result_text, toss_winner):
    if pd.isnull(toss_winner):
        return "bat"
    if pd.isnull(result_text) or "abandoned" in result_text.lower():
        return "bat"
    txt = result_text.lower()
    m = re.search(r"^(\S+)\s+won", txt)
    code = m.group(1).upper() if m else None
    wteam = abbr_mapping.get(code, code)
    if "wickets" in txt:
        return "bat"
    elif "runs" in txt:
        if wteam == toss_winner:
            return "bat"
        else:
            return "field"
    else:
        return "bat"

df_raw['toss_decision'] = df_raw.apply(lambda r: guess_toss_decision(r['Match Winner and Toss'], r['toss_winner']), axis=1)

if 'venue' not in df_raw.columns:
    df_raw['venue'] = "Wankhede Stadium, Mumbai"

fs = pd.ExcelFile("feature_stats.xlsx")
head_to_head_df = fs.parse("Head_to_Head")
venue_h2h_df = fs.parse("Venue_H2H")
toss_stats_df = fs.parse("Toss_Impact")
bat_bowl_df = fs.parse("Batting_Bowling")

if 'teams_pair' not in head_to_head_df.columns:
    head_to_head_df['teams_pair'] = head_to_head_df.apply(lambda r: "-".join(sorted([r['team1'], r['team2']])), axis=1)
if 'teams_pair' not in venue_h2h_df.columns:
    venue_h2h_df['teams_pair'] = venue_h2h_df.apply(lambda r: "-".join(sorted([r['team1'], r['team2']])), axis=1)

def predict_match_ipl24_v3(team1, team2, venue, toss_winner, toss_decision,
                           head_to_head_df, venue_h2h_df, toss_stats_df, bat_bowl_df):
    import random  
    if toss_winner is not None:  # Toss is done
        try:
            toss_factor_team1 = toss_stats_df.loc[toss_stats_df['team'] == team1, 'toss_win_pct'].iloc[0] / 100
        except:
            toss_factor_team1 = 0.5
        try:
            toss_factor_team2 = toss_stats_df.loc[toss_stats_df['team'] == team2, 'toss_win_pct'].iloc[0] / 100
        except:
            toss_factor_team2 = 0.5

        pair_key = "-".join(sorted([team1, team2]))
        hth_row = head_to_head_df[head_to_head_df['teams_pair'] == pair_key]
        if not hth_row.empty:
            hth_row = hth_row.iloc[0]
            sorted_pair = sorted([team1, team2])
            if team1 == sorted_pair[0]:
                hth_factor_team1 = hth_row['team1_win_pct'] / 100
                hth_factor_team2 = hth_row['team2_win_pct'] / 100
            else:
                hth_factor_team1 = hth_row['team2_win_pct'] / 100
                hth_factor_team2 = hth_row['team1_win_pct'] / 100
        else:
            hth_factor_team1 = hth_factor_team2 = 0.5

        vth_row = venue_h2h_df[(venue_h2h_df['venue'] == venue) & (venue_h2h_df['teams_pair'] == pair_key)]
        if not vth_row.empty:
            vth_row = vth_row.iloc[0]
            if team1 == sorted([team1, team2])[0]:
                vth_factor_team1 = vth_row['team1_win_pct'] / 100
                vth_factor_team2 = vth_row['team2_win_pct'] / 100
            else:
                vth_factor_team1 = vth_row['team2_win_pct'] / 100
                vth_factor_team2 = vth_row['team1_win_pct'] / 100
        else:
            vth_factor_team1 = vth_factor_team2 = 0.5

        try:
            row_team1 = bat_bowl_df.loc[bat_bowl_df['team'] == team1].iloc[0]
        except:
            row_team1 = None
        try:
            row_team2 = bat_bowl_df.loc[bat_bowl_df['team'] == team2].iloc[0]
        except:
            row_team2 = None

        if toss_decision.lower() == "bat":
            if toss_winner == team1:
                ctd_team1 = (row_team1['bat_first_win_pct'] if row_team1 is not None else 50) / 100
                ctd_team2 = (row_team2['bowl_first_win_pct'] if row_team2 is not None else 50) / 100
            else:
                ctd_team1 = (row_team1['bowl_first_win_pct'] if row_team1 is not None else 50) / 100
                ctd_team2 = (row_team2['bat_first_win_pct'] if row_team2 is not None else 50) / 100
        else:
            if toss_winner == team1:
                ctd_team1 = (row_team1['bowl_first_win_pct'] if row_team1 is not None else 50) / 100
                ctd_team2 = (row_team2['bat_first_win_pct'] if row_team2 is not None else 50) / 100
            else:
                ctd_team1 = (row_team1['bat_first_win_pct'] if row_team1 is not None else 50) / 100
                ctd_team2 = (row_team2['bowl_first_win_pct'] if row_team2 is not None else 50) / 100

        uncertainty_team1 = random.uniform(0.3, 0.7)
        uncertainty_team2 = random.uniform(0.3, 0.7)
        
        score_team1 = (0.10 * toss_factor_team1 + 0.10 * hth_factor_team1 +
                       0.15 * vth_factor_team1 + 0.05 * ctd_team1 + 0.60 * uncertainty_team1)
        score_team2 = (0.10 * toss_factor_team2 + 0.10 * hth_factor_team2 +
                       0.15 * vth_factor_team2 + 0.05 * ctd_team2 + 0.60 * uncertainty_team2)
    else:
        pair_key = "-".join(sorted([team1, team2]))
        hth_row = head_to_head_df[head_to_head_df['teams_pair'] == pair_key]
        if not hth_row.empty:
            hth_row = hth_row.iloc[0]
            sorted_pair = sorted([team1, team2])
            if team1 == sorted_pair[0]:
                overall_team1 = hth_row['team1_win_pct'] / 100
                overall_team2 = hth_row['team2_win_pct'] / 100
            else:
                overall_team1 = hth_row['team2_win_pct'] / 100
                overall_team2 = hth_row['team1_win_pct'] / 100
        else:
            overall_team1 = overall_team2 = 0.5
        
        vth_row = venue_h2h_df[(venue_h2h_df['venue'] == venue) & (venue_h2h_df['teams_pair'] == pair_key)]
        if not vth_row.empty:
            vth_row = vth_row.iloc[0]
            if team1 == sorted([team1, team2])[0]:
                venue_team1 = vth_row['team1_win_pct'] / 100
                venue_team2 = vth_row['team2_win_pct'] / 100
            else:
                venue_team1 = vth_row['team2_win_pct'] / 100
                venue_team2 = vth_row['team1_win_pct'] / 100
        else:
            venue_team1 = venue_team2 = 0.5
        
        uncertainty_team1 = random.uniform(0.3, 0.7)
        uncertainty_team2 = random.uniform(0.3, 0.7)
        
        score_team1 = (0.15 * overall_team1 + 0.15 * venue_team1 + 0.70 * uncertainty_team1)
        score_team2 = (0.15 * overall_team2 + 0.15 * venue_team2 + 0.70 * uncertainty_team2)
    
    return team1 if score_team1 >= score_team2 else team2

predictions = []
for idx, row in df_raw.iterrows():
    pred = predict_match_ipl24_v3(
        row['team1'], row['team2'], row['venue'],
        row['toss_winner'], row['toss_decision'],
        head_to_head_df, venue_h2h_df, toss_stats_df, bat_bowl_df
    )
    predictions.append(pred)
df_raw['predicted_winner'] = predictions

df_eval = df_raw.dropna(subset=['actual_winner']).copy()
df_eval['correct'] = (df_eval['predicted_winner'] == df_eval['actual_winner']).astype(int)
accuracy = df_eval['correct'].mean() * 100
print(f"\nOverall Group Stage Accuracy: {accuracy:.2f}%")

conf_mat = pd.crosstab(df_eval['actual_winner'], df_eval['predicted_winner'])
print("\nConfusion Matrix (counts):")
print(conf_mat)
plt.figure(figsize=(8,6))
sns.heatmap(conf_mat, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Confusion Matrix for IPL 2024 Group Stage")
plt.xlabel("Predicted Winner")
plt.ylabel("Actual Winner")
plt.tight_layout()
plt.show()

actual_counts = df_eval['actual_winner'].value_counts().sort_index()
pred_counts = df_eval['predicted_winner'].value_counts().sort_index()
win_compare = pd.DataFrame({'Actual Wins': actual_counts, 'Predicted Wins': pred_counts}).fillna(0)
win_compare.plot(kind='bar', figsize=(10,5))
plt.title("Actual vs. Predicted Wins per Team (Group Stage)")
plt.ylabel("Number of Wins")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

pred_win_counts = df_eval['predicted_winner'].value_counts()
points_rows = []
for team in sorted(pred_win_counts.index):
    wins = pred_win_counts[team]
    matches = 14  
    losses = matches - wins
    pts = wins * 2
    points_rows.append([team, matches, wins, losses, pts])
points_table = pd.DataFrame(points_rows, columns=['Team', 'M', 'W', 'L', 'Pts'])
points_table = points_table.sort_values(by=['Pts','Team'], ascending=[False, True]).reset_index(drop=True)
print("\n--- Points Table (No NRR) ---")
print(points_table)

top4 = points_table.head(4)['Team'].tolist()
seed1, seed2, seed3, seed4 = top4
print("\nTop 4 Teams for Playoffs:")
print(points_table.head(4))

playoff_venues = {
    "Qualifier 1": "Narendra Modi Stadium, Ahmedabad",
    "Eliminator": "Narendra Modi Stadium, Ahmedabad",
    "Qualifier 2": "MA Chidambaram Stadium, Chennai",
    "Final": "MA Chidambaram Stadium, Chennai"
}

def playoff_match(t1, t2, venue):
    return predict_match_ipl24_v3(t1, t2, venue, t1, "bat",
                                  head_to_head_df, venue_h2h_df, toss_stats_df, bat_bowl_df)

q1_winner = playoff_match(seed1, seed2, playoff_venues["Qualifier 1"])
q1_loser = seed1 if q1_winner != seed1 else seed2
print(f"\nQualifier 1: {seed1} vs {seed2} at {playoff_venues['Qualifier 1']} -> Winner: {q1_winner}")

elim_winner = playoff_match(seed3, seed4, playoff_venues["Eliminator"])
elim_loser = seed3 if elim_winner != seed3 else seed4
print(f"Eliminator: {seed3} vs {seed4} at {playoff_venues['Eliminator']} -> Winner: {elim_winner}")

q2_winner = playoff_match(q1_loser, elim_winner, playoff_venues["Qualifier 2"])
print(f"Qualifier 2: {q1_loser} vs {elim_winner} at {playoff_venues['Qualifier 2']} -> Winner: {q2_winner}")

final_winner = playoff_match(q1_winner, q2_winner, playoff_venues["Final"])
print(f"Final: {q1_winner} vs {q2_winner} at {playoff_venues['Final']} -> Champion: {final_winner}")

G = nx.DiGraph()
G.add_node(seed1)
G.add_node(seed2)
G.add_node(seed3)
G.add_node(seed4)
G.add_edge(seed1, q1_winner, label="Qualifier 1")
G.add_edge(seed2, q1_winner, label="Qualifier 1")
G.add_edge(seed3, elim_winner, label="Eliminator")
G.add_edge(seed4, elim_winner, label="Eliminator")
G.add_edge(q1_loser, q2_winner, label="Qualifier 2")
G.add_edge(elim_winner, q2_winner, label="Qualifier 2")
G.add_edge(q1_winner, final_winner, label="Final")
G.add_edge(q2_winner, final_winner, label="Final")

plt.figure(figsize=(8,6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=2000, font_size=10, arrowsize=20)
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
plt.title("IPL 2024 Playoff Flowchart")
plt.axis('off')
plt.show()

print("\n--- Final Champion:", final_winner)
print(f"Overall Group Stage Accuracy: {accuracy:.2f}%\n")
```
